```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
df=pd.read_csv("/content/drive/My Drive/heart.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
df.describe()
```

```python
#target distribution
df.target.value_counts()
```

```python
#show correlation matrix
corr=df.corr()
plt.figure(figsize=(18,10))
sns.heatmap(corr,annot=True,)
plt.show()
```

```python
#plot target dist
sns.countplot(x=df.target,hue='sex',data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['not disease','disease'])
plt.title('Target Distribution');
```

```python
sns.countplot(x=df.target,data=df,palette=['#1CA53B','red'])
plt.xticks([0,1],labels=['not disease','disease'])
plt.title('Target Distribution');
```

```python
plt.figure(figsize=(18,10))
sns.countplot(x='age',hue='target',data=df,palette=['#1CA53B','red'])
plt.legend("haven't disease","have disease")
plt.title('Heart disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.distplot(df.age[df['target']==0],bins=30,color="blue")
sns.distplot(df.age[df['target']==1],bins=30,color="purple")
plt.legend()
plt.title('Heart disease distribution for ages')
plt.xlabel('age')
plt.ylabel('frequency')
plt.show()
```
